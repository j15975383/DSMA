```python
!gdown --id "1t5En64DAG_e450RWH1Gb7E6Xbox_-1l9" --output "megathread.csv"
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv("megathread.csv")
df = df.dropna()
df
```

```python
filter = df["0"].str.contains("automatically")
df[filter]
```

```python
filter = df["0"].str.contains("刪除")
df = df[~filter]

df = df.reset_index(drop=True)

df
```

```python
filter = df['0'].str.contains("I am a bot, and this action was performed automatically")
df = df[~filter]

df = df.reset_index(drop=True)

df
```

```python
random.seed(69)
selected = random.sample(range(1, len(df)), 200)
selected
```

```python
selected_text = df.iloc[selected]
selected_text
```

```python
filter = selected_text["0"].str.contains("刪除")
```

```python
data = []
for text in selected_text["0"]:
  print(text)
  column_name = ["text","label"]
  label = input()
  content = [text,label]
  data.append(dict(zip(column_name,content)))
```

```python
test_set = pd.DataFrame(data)
test_set.to_csv("test_set.csv")
```
